### Transfering The dataset into th eprolog knowledge base


```python
def format_key(old_key):
    return "_".join([d.replace("("," ").replace(")"," ").lower().strip() for d in old_key.split()]).replace(" ","_")

visited = set()
symptoms_set =set()
with open('knowledge_base.pl', 'w') as kb_file:
    for index, row in df.iterrows():
        disease = row['Disease']
        symptoms = row[1:]  # Assuming symptoms start from column index 1
        
        for symptom in symptoms:
            if not pd.isna(symptom):
                s= format_key(symptom)
                d= format_key(disease)
                symptoms_set.add(s)
                if (s,d) in visited:
                    continue
                else: visited.add((s,d))
                fact = f"{format_key(symptom)}({format_key(disease)}).\n"
                kb_file.write(fact)
```
